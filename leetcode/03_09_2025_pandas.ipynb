```python
final['attended_exams'] = final['attended_exams'].astype(int)
# Reorder columns to match desired output
final = final[['student_id', 'student_name', 'subject_name', 'attended_exams']]
final
```

```python
def students_and_examinations(students: pd.DataFrame, subjects: pd.DataFrame, examinations: pd.DataFrame) -> pd.DataFrame:
    cnt = examinations.groupby(['student_id','subject_name']).size().reset_index(name='attended_exams')
    
  
    student_exp = students.merge(subjects, how='cross')
    
    final = student_exp.merge(cnt, on=['student_id','subject_name'], how='left').fillna({'attended_exams': 0})
    
    
    final['attended_exams'] = final['attended_exams'].astype(int)
    
 
    final = final.sort_values(by=['student_id','student_name','subject_name'])
    
    return final
```
